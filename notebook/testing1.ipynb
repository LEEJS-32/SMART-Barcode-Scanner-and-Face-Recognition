```python
import cv2
students = {
    "20230001": {"name": "Person1", "seat": "A12"},
    "20230002": {"name": "Person2", "seat": "B05"},
    "20230003": {"name": "Person3", "seat": "C07"},
}

# Load OpenCV's built-in QRCodeDetector (no pyzbar)
detector = cv2.QRCodeDetector()

cap = cv2.VideoCapture(0)
print("ğŸ“· Starting QR scanner... Press 'q' to quit.")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Detect and decode
    data, points, _ = detector.detectAndDecode(frame)

    if data:
        student_id = data.strip()
        print("ğŸŸ¡ QR Code Detected:", student_id)

        if student_id in students:
            student = students[student_id]
            message = f"{student['name']} - Seat {student['seat']}"
            print("âœ… Student Found:", message)
            cv2.putText(frame, message, (30, 50), cv2.FONT_HERSHEY_SIMPLEX,
                        0.8, (0, 255, 0), 2)
        else:
            print("âŒ Student Not Found")
            cv2.putText(frame, "Student Not Found", (30, 50), cv2.FONT_HERSHEY_SIMPLEX,
                        0.8, (0, 0, 255), 2)

    if points is not None:
        pts = points.astype(int).reshape(-1, 2)
        for j in range(len(pts)):
            pt1 = tuple(pts[j])
            pt2 = tuple(pts[(j+1) % len(pts)])
            cv2.line(frame, pt1, pt2, (0, 255, 0), 2)

    cv2.imshow("QR Code Scanner", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import qrcode
import os

students = {
    "20230001": {"name": "Alice Tan", "seat": "A12"},
    "20230002": {"name": "John Lim", "seat": "B05"},
    "20230003": {"name": "Nurul Afiqah", "seat": "C07"},
}

save_dir = "qrcodes"
os.makedirs(save_dir, exist_ok=True)

for student_id in students:
    qr = qrcode.make(student_id)
    qr.save(f"{save_dir}/{student_id}.png")
    print(f"âœ… QR code saved: {save_dir}/{student_id}.png")
```
